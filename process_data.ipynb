```python
import pandas as pd

# Load datasets
df_2016 = pd.read_csv('2016_US_County_Level_Presidential_Results.csv')
df_2020 = pd.read_csv('2020_US_County_Level_Presidential_Results.csv')
df_2024 = pd.read_csv('2024_US_County_Level_Presidential_Results.csv')  # assuming file ends in .csv

# Standardize 2016
df_2016_clean = df_2016.rename(columns={
    'combined_fips': 'county_fips',
    'state_abbr': 'state',
    'per_dem': 'pct_dem',
    'per_gop': 'pct_gop',
    'per_point_diff': 'margin'
})
df_2016_clean['year'] = 2016
df_2016_clean['county_fips'] = df_2016_clean['county_fips'].astype(str).str.zfill(5)

# Standardize 2020
df_2020_clean = df_2020.rename(columns={
    'state_name': 'state',
    'per_dem': 'pct_dem',
    'per_gop': 'pct_gop',
    'per_point_diff': 'margin'
})
df_2020_clean['year'] = 2020
df_2020_clean['county_fips'] = df_2020_clean['county_fips'].astype(str).str.zfill(5)

# Standardize 2024 (same structure as 2020)
df_2024_clean = df_2024.rename(columns={
    'state_name': 'state',
    'per_dem': 'pct_dem',
    'per_gop': 'pct_gop',
    'per_point_diff': 'margin'
})
df_2024_clean['year'] = 2024
df_2024_clean['county_fips'] = df_2024_clean['county_fips'].astype(str).str.zfill(5)

# Select shared columns
columns_to_keep = ['county_fips', 'county_name', 'state', 'votes_gop', 'votes_dem',
                   'total_votes', 'pct_gop', 'pct_dem', 'margin', 'year']

# Keep only aligned columns
df_2016_final = df_2016_clean[columns_to_keep]
df_2020_final = df_2020_clean[columns_to_keep]
df_2024_final = df_2024_clean[columns_to_keep]

# Combine all three years
df_all_years = pd.concat([df_2016_final, df_2020_final, df_2024_final], ignore_index=True)

# Optional: convert margin to float if needed
df_all_years['margin'] = df_all_years['margin'].replace('%', '', regex=True).astype(float)

# Save or preview
df_all_years.to_csv('combined_county_election_results.csv', index=False)
print(df_all_years.head())
```
